```python
from imgaug import augmenters as iaa

def aug_image(image, rotate=(0, 0), shear=(0,0)):

    seq = iaa.Sequential([
        iaa.Affine(
                    rotate= rotate,
                   shear = shear,
                   mode='edge'),

    ])

    image = seq.augment_image(image)
    return image
```
